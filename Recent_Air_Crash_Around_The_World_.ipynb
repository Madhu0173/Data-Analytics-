```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for recent air crashes (hypothetical)
data = {
    'Country': ['USA', 'India', 'China', 'Russia', 'Indonesia', 'Brazil'],
    'Number of Crashes': [15, 8, 12, 10, 6, 9],
    'Fatalities': [400, 200, 300, 180, 150, 250]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Bar Graph - Number of Crashes by Country
plt.figure(figsize=(10, 5))
plt.bar(df['Country'], df['Number of Crashes'], color='skyblue')
plt.title('Air Crashes by Country (Number of Crashes)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Crashes', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie Chart - Percentage of Total Fatalities by Country
plt.figure(figsize=(7, 7))
plt.pie(df['Fatalities'], labels=df['Country'], autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'lightyellow'])
plt.title('Fatalities Distribution by Country', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
